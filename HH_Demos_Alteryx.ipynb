```python
import pandas as pd
import numpy as np
```

```python
hh="Python_HH_Demo_Data.csv"
hh_df = pd.read_csv(hh)
mk="MCCORMICK_ASSIGNMENT.csv"
mk_df = pd.read_csv(mk)
```

```python
hh_df.head()
```

```python
hh_df_new=pd.crosstab(hh_df["MOSAIC_SEGMENT_ID"],hh_df["Value"],values=hh_df["Count"],aggfunc=np.sum)
hh_df_new.head()
```

```python
mk_df.head()
```

```python
mk_df_new = mk_df.drop(['Unnamed: 0','Prob'],axis = 1)
```

```python
mk_df_new_renamed=mk_df_new.rename(columns={"Mosaic":"MOSAIC_SEGMENT_ID"})
mk_df_new_renamed.head()
```

```python
total_df=pd.merge(hh_df_new,mk_df_new_renamed,on = 'MOSAIC_SEGMENT_ID')
```

```python
total_df.head()
```

```python
Seg_total_1=total_df.groupby("Custom_seg")["1 Member HH","2 Member HH","3 Member HH","4 Member HH","5+ Member HH"].sum()
print(Seg_total_1)
```

```python
custom_seg=Seg_total_1.sum(axis=1)
print(custom_seg)
```

```python
totalhh=custom_seg.sum(axis=0)
totalhh
```
