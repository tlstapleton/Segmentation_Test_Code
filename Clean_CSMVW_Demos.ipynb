```python
df=pd.DataFrame()
for x in hh_df['AGE'].unique():
       df=df.join((pd.DataFrame(hh_df[hh_df['AGE'] == '{0}'.format(x)].groupby('MOSAIC_SEGMENT_ID')['INSEQNUM'].count()).rename(columns={'INSEQNUM':'{0}'.format(x)})),how='outer')

for x in hh_df['MARITAL_STATUS'].unique():
       df=df.join((pd.DataFrame(hh_df[hh_df['MARITAL_STATUS'] == '{0}'.format(x)].groupby('MOSAIC_SEGMENT_ID')['INSEQNUM'].count()).rename(columns={'INSEQNUM':'{0}'.format(x)})),how='outer')
    
```
