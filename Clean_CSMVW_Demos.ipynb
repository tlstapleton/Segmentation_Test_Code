```python
import pandas as pd
import numpy as np
hh="Sample_CSMVW.csv"
hh_df = pd.read_csv(hh)
```

```python
#insert the demos to be pulled
demos_header=['AGE','MARITAL_STATUS','INCOME']
```

```python
df=pd.DataFrame()
for demo in demos_header:
    for x in hh_df['{0}'.format(demo)].unique():
           df=df.join((pd.DataFrame(hh_df[hh_df['{0}'.format(demo)] == '{0}'.format(x)].groupby('MOSAIC_SEGMENT_ID')['INSEQNUM'].count()).rename(columns={'INSEQNUM':'{0}'.format(x)})),how='outer')

    
```
